```python
numpyDizisi1 = np.linspace(0,10,20) 
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g")
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"r--")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g-o")
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"-*b")
```

```python
# öncelikle kaç sıra olacağını yaz. Yan yana göstereceksek tek sıra yeterli, o yüzden 1 dedi.
# daha sonra kaç grafik göstereceksem onu yaz (2 dedi.)
# en son hangisini çizeceğimizi söylüyoruz.
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak , şu an 2. grafiği çiziyorum.
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
# Burada 0.9'lar sırasıyla width ve height'ı temsil eder.
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
plt.show()
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.8,0.8])
eksen2 = figure2.add_axes([0.2,0.5,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
# subplots ile  sadece figure değil aynı zamanda eksende döndürdüğünü görürüz. Yani bu aslında bize bir tuple döndürür.
(benimFigure, benimEksen) = plt.subplots()
```

```python
benimEksen
```

```python
# Yukarıda .subplots() ile eksen zaten hazır olarak geldiği için add_axes ile eksen koordinatlarını belirlemeye gerek yok. Direk .plot() ile çizebiliriz.
(benimFigure, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"b")
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout() # 2 grafik arasındaki mesafeyi açar.
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=2)
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend(loc=1)
```

```python
# Eksenin width ve height'ı  plt.figure(figsize=(..,..)) ile ayarlanabilir :
yeniFigur = plt.figure(figsize=(5,5))

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend()
```

```python
# dpi : kaç pixel olacağı yani çözünürlüğü ile ilgili bir değişken.
# dpi'yı genelde grafikle işimiz bitti ve kayıt ediyoruz o zaman kullanırız.
yeniFigur = plt.figure(dpi=150)

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2,label = "numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3")
yeniEksen.legend()
```

```python
yeniFigur.savefig("benimfigure.png", dpi=200)
```
