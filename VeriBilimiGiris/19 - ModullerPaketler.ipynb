```python
import numpy as np
import matplotlib.pyplot as matplot
```

```python
# Ortalama 4000 lira maaş alan ve standart sapması 500 lira olan 1000 tane veri üret :
# numpy.random ile verileri rastgele üret dedik. numpy kütüphanesinin kendi özelliği random.
# numpy.random.normal() ile normal dağılımı ifade eder.
maasListesi = np.random.normal(4000,500,1000)
```

```python
maasListesi
```

```python
np.mean(maasListesi)
```

```python
# matplot daha çok veri görselleştirme için kullanılır.
# maaasListesini bana çizdir ve 50 tane histogram oluştursan yeter:
matplot.hist(maasListesi,50)
matplot.show()
```
