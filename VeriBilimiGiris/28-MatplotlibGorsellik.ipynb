```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,color="#3A95A8",alpha=0.3) # alpha değeri çizgi grafiğine transparanlık katar.
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="green") # color="#C96F23" turuncu rengi verir.
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue", linewidth=3.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="yellow", linewidth=3.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color="#C96F23", linewidth=3.0, linestyle=":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color="r", linewidth=3.0, linestyle="-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color="g", linewidth=3.0, linestyle="--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color="#000000", linewidth=2.0, linestyle="--", marker="o", markerfacecolor="red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color="#000000", linewidth=2.0, linestyle="--", marker="o", markersize=8, markerfacecolor="red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 9, color="#000000", linewidth=2.0, linestyle="--", marker="+", markersize=10)
```
